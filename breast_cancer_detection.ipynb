```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
```

```python
# Target values: 0 = malignant, 1 = benign
df["target"] = df["target"].map({0:"malignant", 1:"benign"})
```

```python
sns.countplot(data=data, x="target")
plt.title("Target distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.drop(columns=['target']).corr(), cmap="coolwarm", annot=False)
plt.title('Correlation Heatmap ')
plt.show()
```

```python
x = df.iloc[:, :-1]
y = df['target']
```

```python
y.value_counts()
```

```python
y = y.map({"malignant":0,"benign":1})
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC()
}
```

```python
models.items()
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_prd = model.predict(x_test) 
    print(f"\n{name}")
    print("Accuracy: ", accuracy_score(y_test, y_prd))
    print(confusion_matrix(y_test, y_prd))
    print(classification_report(y_test, y_prd))
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
rf.score(x_train, y_train)*100, rf.score(x_test, y_test)*100
```

```python
y_prd = rf.predict(x_test)
acc = accuracy_score(y_test, y_prd)
acc
```

```python
cm = confusion_matrix(y_test, y_prd)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Malignant", "Benign"], yticklabels=["Malignant", "Benign"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print("Classification Report:\n", classification_report(y_test, y_prd, target_names=["Malignant", "Benign"]))
```

## save model

```python
import joblib 
joblib.dump(models["Random Forest"], "breast_cancer_model.pkl")
```

## plotting ROC-Curve

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_prob = models["Random Forest"].predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
roc_auc
```

```python
plt.figure()
plt.plot(fpr, tpr, color="darkorange", lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color="navy", linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC - Curve Random Forest")
plt.legend(loc="lower right")
plt.show()
```
